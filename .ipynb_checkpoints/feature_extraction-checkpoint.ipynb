```python
import pandas as pd
import pickle
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pickle.load(open('dataset_level2.pkl','rb'))
# batting_team
# bowling team
# city
# current score
# ball left
# wickets left
# current rr
# last five
```

```python
df
```

```python
df.isnull().sum()
```

```python
df[df['city'].isnull()]['venue'].value_counts()
```

```python
cities = np.where(df['city'].isnull(),df['venue'].str.split().apply(lambda x:x[0]),df['city'])
```

```python
df['city'] = cities
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['venue'],inplace=True)
```

```python
df
```

```python
eligible_cities = df['city'].value_counts()[df['city'].value_counts() > 600].index.tolist()
```

```python
list(eligible_cities)
```

```python
df = df[df['city'].isin(eligible_cities)]
```

```python
df
```

```python
df['current_score'] = df.groupby('match_id')['runs'].cumsum()
```

```python
df
```

```python
df['over'] = df['ball'].apply(lambda x:str(x).split(".")[0])
df['ball_no'] = df['ball'].apply(lambda x:str(x).split(".")[1])
df
```

```python
df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball_no'].astype('int')
df
```

```python
df['balls_left'] = 120 - df['balls_bowled']
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)
df
```

```python
df['player_dismissed'] = df['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
df['player_dismissed'] = df['player_dismissed'].astype('int')
df['player_dismissed'] = df.groupby('match_id')['player_dismissed'].cumsum()
df['wickets_left'] = 10 - df['player_dismissed']
```

```python
df
```

```python
df['crr'] = (df['current_score']*6)/df['balls_bowled']
```

```python
df
```

```python
groups = df.groupby('match_id')

match_ids = df['match_id'].unique()
last_five = []
for id in match_ids:
    last_five.extend(groups.get_group(id).rolling(window=30)['runs'].sum().values.tolist())
```

```python
df['last_five'] = last_five
```

```python
df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five']]
```

```python
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')
```

```python
final_df=final_df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five','runs_x']]
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.isnull().sum()
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df
```

```python
final_df.sample(2)
```

##### spliting data to train and test

```python
# X = final_df.drop(columns=['runs_x'])
# y = final_df['runs_x']
from sklearn.model_selection import train_test_split
# X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)

# Splitting dataset
X = final_df.drop(columns=['runs_x'])
y = final_df['runs_x']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from lightgbm import LGBMRegressor

```

```python
# Preprocessing transformer
trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city'])
], remainder='passthrough')
```

##### Random Forest Model Pipeline


```python
rf_pipe = Pipeline(steps=[
    ('step1', trf),
    ('step2', StandardScaler()),
    ('step3', RandomForestRegressor(n_estimators=500, max_depth=25, max_features="sqrt", n_jobs=-1, random_state=1))
])
```

##### XGBoost Model Pipeline


```python
xgb_pipe = Pipeline(steps=[
    ('step1', trf),
    ('step2', StandardScaler()),
    ('step3', XGBRegressor(n_estimators=1000, learning_rate=0.2, max_depth=12, random_state=1))
])
```

##### Libm Model Pipeline

```python
lgbm_pipe = Pipeline(steps=[
    ('step1', trf),
    ('step2', StandardScaler()),
    ('step3', LGBMRegressor(n_estimators=1000, learning_rate=0.2, max_depth=12, random_state=1))
])
```

```python
# Train all models
rf_pipe.fit(X_train, y_train)
xgb_pipe.fit(X_train, y_train)
lgbm_pipe.fit(X_train, y_train)

# Predictions for test data
rf_pred = rf_pipe.predict(X_test)
xgb_pred = xgb_pipe.predict(X_test)
lgbm_pred = lgbm_pipe.predict(X_test)

# Evaluation
print("\nðŸŒ² Random Forest Performance:")
print("RÂ² Score:", r2_score(y_test, rf_pred))
print("Mean Absolute Error (MAE):", mean_absolute_error(y_test, rf_pred))
print("Mean Squared Error (MSE):", mean_squared_error(y_test, rf_pred))
print("Root Mean Squared Error (RMSE):", np.sqrt(mean_squared_error(y_test, rf_pred)))

print("\nðŸš€ XGBoost Performance:")
print("RÂ² Score:", r2_score(y_test, xgb_pred))
print("Mean Absolute Error (MAE):", mean_absolute_error(y_test, xgb_pred))
print("Mean Squared Error (MSE):", mean_squared_error(y_test, xgb_pred))
print("Root Mean Squared Error (RMSE):", np.sqrt(mean_squared_error(y_test, xgb_pred)))

print("\nâš¡ LightGBM Performance:")
print("RÂ² Score:", r2_score(y_test, lgbm_pred))
print("Mean Absolute Error (MAE):", mean_absolute_error(y_test, lgbm_pred))
print("Mean Squared Error (MSE):", mean_squared_error(y_test, lgbm_pred))
print("Root Mean Squared Error (RMSE):", np.sqrt(mean_squared_error(y_test, lgbm_pred)))
```

```python
# Save both models
pickle.dump(rf_pipe, open('rf_pipe.pkl', 'wb'))
pickle.dump(xgb_pipe, open('xgb_pipe.pkl', 'wb'))
pickle.dump(lgbm_pipe, open('lgbm_pipe.pkl', 'wb'))
```
